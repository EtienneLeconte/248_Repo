```python
url = 'https://api.github.com/events'
response=requests.get(url)
results=response.json()
results
```

```python
data1 = pd.DataFrame(results)
data1.head()
```

```python
to_flatten = ['actor', 'repo','payload','org']
for i in to_flatten:
    flat=pd.DataFrame(dict(data1[i])).T
    flat.columns=[f'{i}_{col}' for col in flat.columns]
    data1=pd.concat([data1,flat],axis=1)
    data1=data1.drop(columns=i)
```
